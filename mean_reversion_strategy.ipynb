```python
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime

# Download historical data for a stock (e.g., Apple)
ticker = 'AAPL'
start_date = '2015-01-01'
end_date = '2023-12-31'

# Load stock data
data = yf.download(ticker, start=start_date, end=end_date)

# Calculate the moving average and standard deviation
window = 20
data['MA20'] = data['Close'].rolling(window=window).mean()
data['StdDev20'] = data['Close'].rolling(window=window).std()

# Define thresholds (e.g., 2 standard deviations)
entry_threshold = 2
exit_threshold = 0

# Create signals
data['LongSignal'] = np.where(data['Close'] < (data['MA20'] - entry_threshold * data['StdDev20']), 1, 0)
data['ShortSignal'] = np.where(data['Close'] > (data['MA20'] + entry_threshold * data['StdDev20']), -1, 0)

# Combine signals into one column
data['Signal'] = data['LongSignal'] + data['ShortSignal']

# Backtest the strategy
initial_capital = 10000
capital = initial_capital
position = 0
data['DailyReturn'] = data['Close'].pct_change()

# Initialize lists to store backtest results
capital_values = [initial_capital]
positions = [position]

for i in range(1, len(data)):
    # If we have a position
    if position != 0:
        # Calculate daily return for the position
        capital += position * data['DailyReturn'][i] * capital
        positions.append(position)
        capital_values.append(capital)
    
    # Update position if there's a signal
    if data['Signal'][i] == 1 and position == 0:  # Go Long
        position = capital / data['Close'][i]
    elif data['Signal'][i] == -1 and position == 0:  # Go Short
        position = -capital / data['Close'][i]
    elif data['Signal'][i] == 0 and position != 0:  # Exit position
        position = 0

# Final result
print(f"Initial Capital: {initial_capital}")
print(f"Final Capital: {capital:.2f}")
print(f"Total Return: {((capital - initial_capital) / initial_capital) * 100:.2f}%")

# Plot the equity curve
plt.figure(figsize=(10, 6))
plt.plot(data.index[:len(capital_values)], capital_values, label='Strategy Equity Curve')
plt.title(f'Mean Reversion Strategy Backtest ({ticker})')
plt.xlabel('Date')
plt.ylabel('Capital ($)')
plt.legend()
plt.show()
```

```python
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Download Apple stock data from Yahoo Finance
ticker = 'AAPL'
start_date = '2015-01-01'
end_date = '2023-12-31'

data = yf.download(ticker, start=start_date, end=end_date)

# Calculate daily returns
data['DailyReturn'] = data['Close'].pct_change()

# Calculate cumulative returns
data['CumulativeReturn'] = (1 + data['DailyReturn']).cumprod()

# Calculate the running maximum of the cumulative returns (peak value at any time)
data['CumulativeMax'] = data['CumulativeReturn'].cummax()

# Calculate the drawdown (percentage decline from the peak)
data['Drawdown'] = (data['CumulativeReturn'] - data['CumulativeMax']) / data['CumulativeMax']

# Plot the Drawdown
plt.figure(figsize=(10, 6))
plt.plot(data.index, data['Drawdown'], label='Drawdown', color='red')
plt.fill_between(data.index, data['Drawdown'], color='red', alpha=0.3)  # Fill the area under the curve
plt.title(f'Drawdown for {ticker} Stock (2015-2023)')
plt.xlabel('Date')
plt.ylabel('Drawdown (%)')
plt.grid(True)
plt.legend(loc='best')
plt.show()
```

```python
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Download Apple stock data from Yahoo Finance
ticker = 'AAPL'
start_date = '2021-01-01'
end_date = '2023-12-31'

data = yf.download(ticker, start=start_date, end=end_date)

# Parameters
window = 45  # Window for moving average and standard deviation
std_dev_factor = 2  # Number of standard deviations for entry signal

# Calculate the moving average and rolling standard deviation
data['MA20'] = data['Close'].rolling(window=window).mean()
data['StdDev20'] = data['Close'].rolling(window=window).std()

# Entry signal: Go long when price is 2 standard deviations below the moving average
data['LongSignal'] = np.where(data['Close'] < (data['MA20'] - std_dev_factor * data['StdDev20']), 1, 0)

# Exit signal: Sell when the price hits the moving average
data['ExitSignal'] = np.where(data['Close'] >= data['MA20'], 1, 0)

# Backtest the strategy
initial_capital = 10000
capital = initial_capital
position = 0
data['DailyReturn'] = data['Close'].pct_change()

# Track capital and position for each day
capital_values = [initial_capital]
positions = [position]
for i in range(1, len(data)):
    # If we have a position (long)
    if position != 0:
        capital += position * data['DailyReturn'][i] * capital  # Update capital based on price movement
        capital_values.append(capital)
        positions.append(position)

    # Update position if there's a signal to buy (go long) or sell (exit)
    if data['LongSignal'][i] == 1 and position == 0:  # Go Long if price is below 2 standard deviations
        position = capital / data['Close'][i]
    elif data['ExitSignal'][i] == 1 and position != 0:  # Sell if price hits the moving average
        position = 0  # Exit position
        capital_values.append(capital)
        positions.append(position)

# Final result
print(f"Initial Capital: {initial_capital}")
print(f"Final Capital: {capital:.2f}")
print(f"Total Return: {((capital - initial_capital) / initial_capital) * 100:.2f}%")

# Plot the equity curve
plt.figure(figsize=(10, 6))
plt.plot(data.index[:len(capital_values)], capital_values, label='Strategy Equity Curve')
plt.title(f'Mean Reversion Strategy Backtest ({ticker})')
plt.xlabel('Date')
plt.ylabel('Capital ($)')
plt.legend()
plt.grid(True)
plt.show()

# Plot price and moving average with entry and exit signals
plt.figure(figsize=(10, 6))
plt.plot(data['Close'], label='Close Price', alpha=0.7)
plt.plot(data['MA20'], label='20-Day Moving Average', color='orange')
plt.scatter(data.index[data['LongSignal'] == 1], data['Close'][data['LongSignal'] == 1], marker='^', color='g', label='Buy Signal', alpha=1)
plt.scatter(data.index[data['ExitSignal'] == 1], data['Close'][data['ExitSignal'] == 1], marker='v', color='r', label='Sell Signal', alpha=1)
plt.title(f'Mean Reversion Strategy Signals ({ticker})')
plt.xlabel('Date')
plt.ylabel('Price ($)')
plt.legend()
plt.grid(True)
plt.show()
```
