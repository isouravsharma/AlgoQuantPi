```python
## Cummulative Returns

# Cummulative Returns = (Ending Price - Starting Price)/Starting Price
```

# Libraries we need 
- Pandas 
- yfinance 
- matplotlib

```python
# Import Libraries 
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ticker = 'TSLA'

data = yf.download(tickers=ticker, interval='1d')
```

```python
data
```

```python
# calulate the Daily returns 
data['Daily Return'] = data['Adj Close'].pct_change()
```

```python
# Cummulative Return
data['Cumulative Return'] = (1 + data['Daily Return']).cumprod()-1
```

```python
data.tail()
```

```python
# Visualize 

plt.figure(figsize=(10, 6))
plt.plot(data.index, data['Cumulative Return'], label = 'Cumulative Return')
plt.xlabel('Years')
plt.ylabel('Cumulative Return')
plt.grid(True)
plt.legend()
plt.show()
```
